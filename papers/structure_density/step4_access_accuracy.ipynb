```python
# Local path, change this.
yellow_follow = 'C:/Users/caspe/Desktop/yellow/lib/'

import sys; sys.path.append(yellow_follow) 
import sqlite3
import pandas as pd
from os import path
from glob import glob
```

```python
folder = "C:/Users/caspe/Desktop/Paper_2_StructuralVolume/"
subfolder = "results_volume/"

results = glob(folder + subfolder + "*.sqlite")
target = "volume"

for result in results:
    filename = os.path.splitext(path.basename(result))[0]
    grid_name = f"grid_{'_'.join(filename.split('_')[1:])}"

    in_path = folder + f"{subfolder}{filename}.sqlite"
    db_cnx = sqlite3.connect(in_path)
    df = pd.read_sql_query(f"SELECT * FROM '{filename}';", db_cnx)

    df_pt = pd.DataFrame({ "pred": df[grid_name].values, "target": df[target].values })
    df_above0 = df_pt[df_pt["target"] > 0]

    mae = abs(df_above0["pred"] - df_above0["target"]).mean()
    meae = abs(df_above0["pred"] - df_above0["target"]).median()

    print(f"    Test area: {filename}")
    print(f"    Mean Absolute Error (MAE):                {ml_utils.pad(str(round(mae, 3)), 5, 3)}")
    print(f"    Median Absolute Error (MeAE):             {ml_utils.pad(str(round(meae, 3)), 5, 3)}")
    print("")

```
